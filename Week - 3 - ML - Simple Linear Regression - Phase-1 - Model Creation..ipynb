```python
import pandas as pd  #Pandas Supports Tables
```

```python
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset 
```

```python
#Input\Output Split Method
```

```python
independent=dataset[["YearsExperience"]] #Independent - Input in the dataset
independent 
```

```python
dependent=dataset[["Salary"]] #Dependent - Output in the dataset
dependent
```

```python
#Training set and Test Set Split Method
```

```python
from sklearn.model_selection import train_test_split  #Sklearn Supports Machine Learning
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression  #These first two lines are called as a procedure
regressor=LinearRegression()
regressor.fit(X_train,y_train) #Fit - Called Function
```

```python
weight=regressor.coef_  #How to see the calculated weight
weight
```

```python
bais=regressor.intercept_ #How to see bais
bais
```

```python
y_pred=regressor.predict(X_test) #Evaluation Metric - For Model Evaluation
```

```python
from sklearn.metrics import r2_score  #R2 Value Identification
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle #How to save model
filename="finalized_model_linear.sav" #File name must be in .sav in the last
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb')) #Checking a saved model, whether working properly 
result=loaded_model.predict([[15]])
```

```python
#ML - Regression - Decision Tree - Algorithm
```

```python
import pandas as pd #Pandas is a library
```

```python
dataset=pd.read_csv("50_Startups.csv") #Dataset file name
```

```python
dataset #Function call
```

```python
dataset=pd.get_dummies(dataset,drop_first=True) #Categorical data converted into numerical data. This technique is called One-Hot Encoding.
```

```python
dataset
```

```python
dataset.columns #Input dataset and output dataset - split
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']] #Independent - Input dataset
```

```python
dependent=dataset[["Profit"]] #Dependent - Output dataset
```

```python
#Split - Training set and test set
```

```python
from sklearn.model_selection import train_test_split  #Sklearn Supports Machine Learning
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#Model creation
```

```python
from sklearn.tree import DecisionTreeRegressor  #These first two lines are called as a procedure
regressor=DecisionTreeRegressor(criterion='squared_error',splitter='best') #DecisionTreeRegressor - Algorithm. 'Squared_error' - One of the parameters.
regressor=regressor.fit(X_train,y_train) #Fit - Called Function. X-input, Y-output
```

```python
import matplotlib.pyplot as plt #Matplotlib is a library, it uses for graph purposes.
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(X_test) #Evaluation Metric - For Model Evaluation
```

```python
from sklearn.metrics import r2_score  #R2 Value Identification
r_score=r2_score(y_test,y_pred)
```
